Boolean mask showing which rows satisfy a condition

```python
ww = (df1['zHD'] > 0.01) # including only rows where zHD > 0.01
df1_ww = df1[ww]
print(df1_ww)
```

Instead of keeping the entire DataFrame, you are extracting only specific  columns (array)
