# File Handling 

```python
import numpy as np
```

To create a new file in Python, use the open() method


```python
# Write to the file
f= open("my_file.txt", "w") # Instead of f you can use file as variable name
f.write("Cosmology prerequisites \n Are you ready ?" )
f.close()
```

```python
# Read from the file
f = open("my_file.txt", "r")
print(f.read())  # reading is stored in content variable
f.close()
```

Redo it 

```python
# Write to the file
f= open("my_file.txt", "w") # Instead of f you can use file as variable name
f.write("Cosmology prerequisites \n Are you ready ?")
f.close()
```

```python
# Read from the file
f = open("my_file.txt", "r")
print(f.readline())  # readline() reads only one line
f.close()

```

  r+ mode is used for both reading and writing

```python
f = open("my_file.txt", "r+")  
f.write("Cosmology prerequisites \n Are you ready ?" )
print(f.read()) #read() means read the entire content of the file
f.close()
# give no output
```

```python
f = open("my_file.txt", "r+")   # always read and write from beginning
print(f.read())
f.write("Cosmology prerequisites" )
f.close()
```

```python
data= np.genfromtxt("/Users/anilkandel/Documents/Anil LCDM/data/HzTable_MM_BC03.dat", delimiter=",")
z_data=data[:,0]
H_data=data[:,1] # This store  H data of all row & 2nd column
print(z_data)
print(H_data)  # This store z, H data in separate arrays
```

```python
data = "/Users/anilkandel/Documents/Anil LCDM/data/HzTable_MM_BC03.dat"

# Load data. (unpack=True) Split columns automatically into separate arrays
z, Hz, errHz = np.genfromtxt(data, comments='#', usecols=(0,1,2), unpack=True, delimiter=',')

# Print arrays
print("z:", z)
print("Hz:", Hz)
print("errHz:", errHz)
```
