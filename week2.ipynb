```python
import matplotlib.pyplot as plt # plt is an alias for matplotlib.pyplot
import numpy as np
```

```python
t=np.arange(-10,10,0.001) # Takes each value in t and computes the sine of it.
y=np.sin(t)
plt.xlabel("t")
plt.ylabel("sin(x)") 
plt.plot(t,y)
```

```python
t=np.arange(-10,10,0.001) # Takes each value in t and computes the sine of it.
y=np.sin(t)
plt.xlabel("t")
plt.ylabel("sin(x)") 
plt.title(" sin(x) vs x")
plt.plot(t,y, color='red') 
```

# Add grid and legend( Multiple plots)

```python
t=np.arange(-10,10,0.001) # Takes each value in t and computes the sine of it.
y1=np.sin(t)
y2=np.cos(t)
plt.xlabel("t")
plt.ylabel("sin(x)") 
plt.plot(t,y1, color='red') 
plt.plot(t,y2, color='blue')
plt.legend(['sin(t)', 'cos(t)'])

```

```python
t=np.arange(-10,10,0.001) # Takes each value in t and computes the sine of it.
y1=np.sin(t)
y2=np.cos(t)
plt.xlabel("t")
plt.ylabel("sin(x)") 
plt.plot(t,y1, color='red') 
plt.plot(t,y2, color='blue')
plt.legend(['sin(t)', 'cos(t)'], loc='upper right')
```

```python
t = np.arange(0, 10, 0.001)
y1 = np.sin(t)
y2 = np.cos(t)

# Plot lines
plt.plot(t, y1, color='red')
plt.plot(t, y2, color='blue')

# Labels
plt.xlabel("t")
plt.ylabel("function")

# Legend outside the plot
plt.legend(['sin(t)', 'cos(t)'], loc='upper left', bbox_to_anchor=(1.1, 1)) # (0,0  bottom-left of the axes) &  (1,1  top-right of the axes)

# Adjust layout to make room for the legend
plt.tight_layout()

# Show plot
plt.show()
```

# Save plot 

```python
t=np.arange(-10,10,0.001) # Takes each value in t and computes the sine of it.
y1=np.sin(t)
y2=np.cos(t)
plt.xlabel("t")
plt.ylabel("sin(x)") 
plt.plot(t,y1, color='red') 
plt.plot(t,y2, color='blue')
plt.legend(['sin(t)', 'cos(t)'], loc='upper right')
plt.savefig('Sample_Outplot.png')
plt.savefig('Sample_Outplot.pdf')

```

# Pie Chart

```python
import matplotlib.pyplot as plt

x = (4.3, 27.0, 68.7)
labels = ['Baronic Matter', 'Dark Matter', 'Dark energy'] # Add label
plt.pie(x, labels=labels)
```

```python
x = [4.3, 27.0, 68.71]
labels = ['Baryonic Matter', 'Dark Matter', 'Dark Energy']
colors = ['gold', 'gray', 'skyblue']  # custom colors
plt.title('Composition of the Universe')
plt.pie(x, labels=labels, colors=colors, autopct='%1.1f%%') # autopct (auto percentage) show percentage in 1 decimal place
plt.tight_layout()
plt.show()
```

Add Legend

```python
y = [4.3, 27.0, 68.71]
colors = ['gold', 'gray', 'skyblue']  # custom colors
plt.title('Composition of the Universe')
plt.pie(y, colors=colors, autopct='%1.1f%%') # autopct (auto percentage) show percentage in 1 decimal place
plt.legend(['Baryonic Matter', 'Dark Matter', 'Dark Energy'], loc='upper right', bbox_to_anchor=(1.3, 1)) 
plt.tight_layout()
plt.show()
```

subplots

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 9))

plt.subplot(1, 2, 1)  # I will get 2 plots in 1 row and 2 columns this is for over all 2 plots representation
plt.pie([4.3, 27.0, 68.71], colors=colors, autopct='%1.1f%%')
plt.title('Interacting Dark Energy Model')

plt.subplot(1, 2, 2)
plt.pie([4.0, 26, 69.0], colors=colors, autopct='%1.1f%%')
plt.title('Non-Interacting Dark Energy Model')
plt.legend(['Baryonic Matter', 'Dark Matter', 'Dark Energy'], loc='center left', bbox_to_anchor=(1.0, 1))


plt.tight_layout()
plt.savefig('Dark_Energy_Models.pdf')
plt.show()
```
